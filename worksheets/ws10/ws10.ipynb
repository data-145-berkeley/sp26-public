## 1. Simple Linear Regression as a Canonical Linear Model

In this problem you will derive the $t$-test for $H_0: \beta_1 = 0$ in simple linear regression by applying the canonical linear model machinery from Lecture 24. In both parts below the fixed predictor values $x_1, \ldots, x_n$ are given constants (not all equal to zero, and in part (b) not all equal), and $Y_1, \ldots, Y_n$ are independent.

For each part, you will:

1. Write the model in vector form and identify the model subspace $\Theta$ and null subspace $\Theta_0$.
2. Determine the block dimensions $(d_0, d_1, d_r)$.
3. Find the orthonormal basis vectors $\mathbf q_0$ and $\mathbf q_1$.
4. Compute the canonical coordinates $Z_0$ and $Z_1$ in terms of the data.
5. Compute the residual magnitude $\|Z_r\|^2$ using the orthogonal decomposition $\|Y\|^2 = Z_0^2 + Z_1^2 + \|Z_r\|^2$, and verify it equals the residual sum of squares from the OLS fit.
6. Build the $t$-statistic and identify its null distribution.

Notation throughout: $\mathbf x = (x_1, \ldots, x_n)^T$, $\mathbf 1 = (1, \ldots, 1)^T$, $\mathbf Y = (Y_1, \ldots, Y_n)^T$, $\bar x = \frac{1}{n}\sum x_i$, $\bar Y = \frac{1}{n}\sum Y_i$, $S_{xx} = \sum (x_i - \bar x)^2$, $S_{xy} = \sum (x_i - \bar x)(Y_i - \bar Y)$, $S_{YY} = \sum (Y_i - \bar Y)^2$.

**(a.4)** Find the orthonormal basis vector $\mathbf q_1$ (a unit vector in $\Theta$ orthogonal to $\Theta_0$). There is no $\mathbf q_0$ in this part — explain briefly why.

**(a.5)** Compute $Z_1 = \mathbf q_1^T \mathbf Y$ in terms of $x_i$ and $Y_i$. What is the distribution of $Z_1$ under $H_0$? Under $H_1$?

**(a.6)** Using the orthogonal decomposition $\|\mathbf Y\|^2 = Z_1^2 + \|Z_r\|^2$, derive a formula for $\|Z_r\|^2$ in terms of $\sum Y_i^2$, $\sum x_i^2$, and $\sum x_i Y_i$.
