## 2. RCT: Covariance of Treatment and Control Means
This is the step left for you in Lecture 27. Notation recap:
- Each unit (or individual) in the RCT has a potential treatment outcome and a potential control outcome.
- potential treatment outcomes $t_1, t_2, \ldots, t_N$, mean $\bar{t}$, variance $\sigma_t^2$
- potential control outcomes $c_1, c_2, \ldots, c_N$, mean $\bar{c}$, variance $\sigma_c^2$
- covariance of treatment and control outcomes, where $(t_i, c_i)$ is the pair of outcomes of unit $i$: $cov(t, c) = \displaystyle \frac{1}{N}\sum_{i=1}^N (t_i - \bar{t})(c_i - \bar{c})$.
- Observed treatment outocmes: $X_1, X_2, \ldots, X_n$ is a simple random sample from the $N$ potential treatment outcomes. The average is $\bar{X}$.
- Observed control outcomes: $Y_1, Y_2, \ldots, Y_m$ is a simple random sample of potential control outcomes from the $N-n$ indices not selected for treatment above. The average is $\bar{Y}$.

As discussed in class, $\bar{X} - \bar{Y}$ is an unbiased estimator of the average treatment effect $\bar{t}-\bar{c}$. The goal of this exercise is to prove the following fact used in the calculation of $Var(\bar{X} - \bar{Y})$:
$$
Cov(\bar{X}, \bar{Y}) = -\frac{1}{N-1}cov(t, c)
$$
