Let $H_0$ and $H_1$ be two simple hypotheses. Let $f_0$ be the likelihood of the sample $\mathbb{X}$ under the null hypothesis $H_0$, and let $f_1$ be its likelihood under the alternative hypothesis $H_1$. Let $LR(\mathbb{X})$ be the likelihood ratio of the data:
$$
LR(\mathbb{X}) = \frac{f_1(\mathbb{X})}{f_0(\mathbb{X})}
$$

Let $T$ be the likelihood ratio test that rejects $H_0$ if $LR(\mathbb{X}) > c$ for some threshold $c > 0$. Let $\alpha$ be the level of $T$.
