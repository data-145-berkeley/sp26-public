## [RECAP FROM LEC 15] The Generalized Likelihood Ratio Test

### Idea

Compare the maximum log-likelihood (at $\hat\theta_n$) to the log-likelihood at the null value $\theta_0$. If $H_0$ is true, these shouldn't be too different.

### Test statistic

$$\text{GLRT statistic} = 2[\ell_n(\hat\theta_n; X) - \ell_n(\theta_0; X)]$$

This is always $\geq 0$ (since $\hat\theta_n$ maximizes $\ell_n$). Reject $H_0$ if it is large.

### Derivation of the null distribution

Taylor-expand $\ell_n(\theta)$ around $\hat\theta_n$, using the fact that $\ell_n'(\hat\theta_n) = 0$:

$$\ell_n(\theta_0; X) - \ell_n(\hat\theta_n; X) \approx \frac{1}{2}\ell_n''(\hat\theta_n; X)(\theta_0 - \hat\theta_n)^2$$

Since $-\ell_n''(\hat\theta_n; X) \approx nI(\theta_0)$ by the law of large numbers, we get:

$$2[\ell_n(\hat\theta_n; X) - \ell_n(\theta_0; X)] \approx nI(\theta_0)(\hat\theta_n - \theta_0)^2$$

Under $H_0$: $\sqrt{nI(\theta_0)}(\hat\theta_n - \theta_0) \overset{d}{\to} N(0,1)$, so squaring:

$$2[\ell_n(\hat\theta_n; X) - \ell_n(\theta_0; X)] \overset{d}{\to} \chi^2_1$$

Reject $H_0$ if $2[\ell_n(\hat\theta_n; X) - \ell_n(\theta_0; X)] \geq \chi^2_{1,\alpha}$.

### Poisson example

For $\text{Pois}(\lambda)$ data with log-likelihood $\ell_n(\lambda) = n\bar{X}\log\lambda - n\lambda + C$, the GLRT statistic is:

$$2[\ell_n(\hat\lambda) - \ell_n(\lambda_0)] = 2\bigl[n\bar{X}\log\bar{X} - n\bar{X} - n\bar{X}\log\lambda_0 + n\lambda_0\bigr] = 2n\left[\bar{X}\log\frac{\bar{X}}{\lambda_0} - (\bar{X} - \lambda_0)\right]$$

Note that each term in the brackets is non-negative: $\bar{X}\log(\bar{X}/\lambda_0) \geq \bar{X} - \lambda_0$ by Jensen's inequality (with equality iff $\bar{X} = \lambda_0$).

## 1. Re-writing the Generalized Likelihood Ratio Statistic

Suppose we are testing the hypotheses $H_0: \theta = \theta_0$ versus $H_1: \theta \neq \theta_0$.

Start by defining the *generalized likelihood ratio*
$$
\Lambda ~ = ~ \frac{\displaystyle\max_{\text{all }\theta} \text{likelihood of the data under } \theta}{\text{likelihood of the data under } \theta_0} ~ = ~ \frac{\text{Lik}(\hat{\theta}; X)}{\text{Lik}(\theta_0; X)}
$$
The numerator is the value of the likelihood calculated at $\hat{\theta}$, the MLE of $\theta$. Hence the value of the ratio is at least $1$, and large values favor $H_1$. We should reject $H_0$ when $\Lambda$ is large.

That's the same as rejecting $H_0$ when $\log(\Lambda)$ is large.
$$
\log(\Lambda) ~ = ~ \ell(\hat{\theta}; X) - \ell(\theta_0; X) ~ \ge ~ 0 ~~~ \text{since } \Lambda \ge 1
$$

By the derivation in Lecture 15, $2\log(\Lambda)$ is approximately $\chi_1^2$ under $H_0$ if the sample size is large. So the test statistic is
$$
\text{GLRT statistic} ~ = ~ 2\log(\Lambda)
$$
We reject $H_0$ for large values $2\log(\Lambda)$. Assuming our sample size is big enough, we can use the $\chi_1^2$ distribution and our choice of level to determine an approximate cutoff for "large".

## 2. Generalized Likelihood Ratio Test with Multiple Parameters

The test above has a commonly used extension in settings where there are multiple parameters. We will focus on the multinomial setting with $c$ categories, and we will start by testing a simple null hypothesis.

$H_0$: The sample is i.i.d from a completely specified categorical distribution on the $c$ categories.

$H_1$: The sample is i.i.d from some different distribution on the $c$ categories.

Let the proportions specified in the null distribution be $\theta_{0, i}$ for $1 \le i \le c$. Then $\displaystyle \sum_{i=1}^c \theta_{0,i} = 1$. So there are only $c-1$ free parameters. The remaining parameter is determined by the constraint that the parameters must sum to $1$.

#### Theorem (which we will not prove)
Suppose our data are $X_1,X_2,...,X_n$ i.i.d $f_\theta$ where $\theta$ is a vector of parameters. And suppose we are testing $H_0: \theta = \theta_0$ versus $H_1: \theta \neq \theta_0$. Assume nice conditions on $f$ and the set of possible $\theta$. 

Then under $H_0$, $2\log{\Lambda} \stackrel{d}{\to} \chi^2_k$ as $n \to \infty$, where $k$ is the number of free parameters.

In the case of our completely specified null distribution in the multinomial model, the number of free parameters is $c-1$, one fewer than the number of categories. When the sample is large, the multinomial proportions tend to a multivariate normal that sums to 1. So its dimension is $c-1$ instead of $c$. So it makes sense that the degrees of freedom of the chi-squared statistic should be $c-1$. 

### Calculating the Statistic

Let the probabilities in the null distribution be $\theta_{0, i}$ for $1 \le i \le c$. Then $\sum_{i=1}^c \theta_{0,i} = 1$. 

For $i = 1,2...,c$, let $O_i$ be the observed frequency (that is, count) in Category $i$. The MLE of $\theta_i$, the probability of a draw being in Category $i$, is the observed proportion in that category. That is, $\hat{\theta}_i = O_i/n$.

**Note:** This result needs an argument though it looks as though we already know it. The reason is that the likelihood must be maximized over all $c$ parameters simultaneously. This can be done using a Lagrange multiplier since the sum of the parameters has to be 1. We won't do that calculation. The resulting estimates are exactly what we would want them to be.

Let $E_i$ be the expected frequency in Category $i$ under $H_0$. Then $E_i = n\theta_{0, i}$.

The generalized likelihood ratio is

$$
\begin{align*}
\Lambda &= \frac
{\frac{n!}{O_1!O_2!\cdots O_c!}\prod_{i=1}^c\hat{\theta}^{O_i}}
{\frac{n!}{O_1!O_2!\cdots O_c!}\prod_{i=1}^c\theta_{0,i}^{O_i}}
\\ \\
&= \frac
{\prod_{i=1}^c(O_i/n)^{O_i}}
{\prod_{i=1}^c\theta_{0,i}^{O_i} }
\\ \\
&= \frac{\prod_{i=1}^c {O_i}^{O_i}}{\prod_{i=1}^{c}{E_i}^{O_i}}
\\ \\
&=\prod_{i=1}^{c}\left(\frac{O_i}{E_i} \right)^{O_i}
\end{align*}
$$
So
$$
2\log\Lambda ~ = ~ 2\sum_{i=1}^{c}O_i \log\frac{O_i}{E_i}
$$

By the theorem above, the null distribution of this statistic converges to $\chi_{c-1}^2$ as the sample size gets large.

#### Important Note
You *must* use frequencies (that is, counts) in this calculation. If you replace them with proportions, you will be off by a factor of $1/n$.

### Example: Is the Roulette Wheel Fair?

Suppose we spin a Nevada roulette wheel and the winning pocket is red 192 times, black 165 times, and green 23 times. Is the wheel fair?

This is the multinomial model with $H_0$ specifying the probabilities of red, black, and green as 18/38, 18/38, and 2/38 respectively. 

This distribution is completely specified. There is nothing to estimate. The question is whether the data look like random draws from this distribution.

```python
observed = np.array([192, 165, 23])
num_spins = sum(observed)
expected = num_spins * np.array([18/38, 18/38, 2/38])
stats.chisquare(observed, expected)
```

The impressively named `Power_divergenceResult` is an object that has the two attributes `statistic` and `pvalue`. The value of the statistic is 2.5, which is pretty close to the expectation of 2 under the null. That explains the large $p$-value.

Check that the statistic is what our formula says it should be:

```python
statistic = 2 * sum(observed * np.log(observed / expected))
statistic
```

This doesn't exactly agree with Python's statistic. Hold that thought. We'll soon see what Python has done.

But it's close enough. Check that the $p$-value is what we say it should be:

```python
1 - stats.chi2.cdf(2.5, df = 2)
```

### Chi-Squared Statistic or Total Variation Distance?

In Data 8 we used the total variation distance (TVD) to assess the difference between two categorical distributions. Unlike the generalized likelihood ratio statistic, the TVD works every time even if the sample size is small. However, since the null distribution of the TVD doesn't have a nice formula, we have to simulate it every time. 

The chi-squared test is computationally simpler but it requires large samples. 

### Chi-Squared Statistic or Kolmogorov-Smirnov Distance?

The Kolmogorov-Smirnov statistic we developed is used to test whether the data are from a continuous, increasing cdf. The chi-squared test we have developed here is for whether categorical data are from a multinomial distribution.

## 3. Pearson's Approximation of the $\chi^2$ Statistic

Pearson showed that under the null hypothesis, the statistic above can be approximated by the following expression when the sample is large. This approximation was useful for conducting $\chi^2$ tests in the era before computers. It is commonly taught in introductory statistics classes.

$$
2\log{\Lambda} ~=~ 2\sum_{i=1}^{c}{O_i}\log\frac{O_i}{E_i} ~\approx~ \sum_{i=1}^c \frac{ (O_i - E_i)^2 }{E_i} ~~~~~
\text{under } H_0 \text{ if the sample is large}
$$

Let's compute this for our roulette data.

```python
pearson_chisquare = sum ( (observed - expected)**2 / expected)
pearson_chisquare
```

That's the statistic we got by using `stats.chisquare`, and that's why the documentation of `stats.chisquare` says it performs Pearson's chi-squared test.

To get a sense of why Pearson's approximation holds, let $D_i = O_i - E_i$. Notice that $\sum_i D_i = \sum_i (O_i - E_i) = \sum_i O_i - \sum_i E_i = n - n = 0$. Now
$$
2\log(\Lambda) ~ = ~ 2 \sum_{i=1}^c (E_i + D_i) \log \frac{E_i + D_i}{E_i} ~ = ~ 2 \sum_{i=1}^c (E_i + D_i) \log \left( 1 + \frac{D_i}{E_i} \right)
$$

If $H_0$ is true and the sample is large, you can expect $O_i$ to be close to $E_i$, that is, you can expect $D_i$ to be small relative to $E_i$. So use the approximation $\log(1 + x) \sim x - \displaystyle \frac{x^2}{2}$ when $x$ is small. This is a little more careful that the approximation $\log(1+x) \sim x$ when $x$ is small, which you might recall from your probability class.

Under $H_0$, if the sample is large,
$$
\begin{align*}
2 \log(\Lambda)~ &\approx ~  2 \sum_{i=1}^c (E_i + D_i)\left( \frac{D_i}{E_i} - \frac{D_i^2}{2E_i^2} \right) \\ \\
&\approx ~ 2 \sum_{i=1}^c \left(D_i - \frac{D_i^2}{2E_i} + \frac{D_i^2}{E_i} - \frac{D_i^3}{2E_i^2} \right) \\ \\
&\approx ~ 2 \sum_{i=1}^c D_i + \sum_{i=1}^c \frac{D_i^2}{E_i}
\end{align*}
$$
ignoring terms of order greater than $D_i^2$. Since $\sum_i D_i = 0$, we have
$$
2 \log(\Lambda) ~ \approx ~ \sum_{i=1}^c \frac{D_i^2}{E_i} ~ = ~ \sum_{i=1}^c \frac{ (O_i - E_i)^2 }{E_i}
$$
if $H_0$ is true and the sample is large.

## 4. Goodness of Fit with an Estimated Parameter

The chi-squared test can be modified to test whether categorical data are drawn from a distribution family with an unknown parameter. For example, for i.i.d. $X_1, X_2, \ldots, X_n$ with values 0 through 6, we can test if they are binomial $(6, p)$ variables for some $p$.

This test works exactly like the one we just performed, except that the number of free parameters drops by 1, leading to a further drop in the degrees of freedom. We have to start by estimating $p$. The degrees of freedom become $7 - 1 - 1$, where the first $-1$ is because the proportions must add up to $1$, and the second $-1$ is because we are constrained to have a particular estimate of $p$. 

This is not a rigorous argument by any means, and it can fail to convince. Since the math involved is out of our scope, we will simply observe what happens by simulation.

### Example: Do the Data Follow a Poisson Distribution?

Suppose we want to know whether $X_1, X_2, \dots X_n$ is an i.i.d. Poisson ($\lambda$) sample, where $\lambda$ is unknown. In your worksheet, you will perform a chi-squared test in the following steps.

**Step 1:** Estimate $\lambda$ from the data. The MLE of the Poisson mean is $\hat{\lambda} = \bar{X}_n$, the sample mean.

**Step 2:** The support of the Poisson distribution is infinite. Truncate the sampled data to create $c$ categories $0, 1, \dots c-1$, by sweeping the last few cells into one. To avoid poor approximations in the tail, it's a good idea to truncate so that there are still some observations in the last cell.

**Step 3:** For $0 \le i \le c-1$, calculate $E_i$, the expected count in Category $i$ under the Poisson ($\hat{\lambda}$) distribution. For category $c-1$, use the Poisson $(\hat{\lambda})$ distribution to find $E_{c-1}$. The expected proportion in that cell is $P(X \geq c-1) = 1-P(X \leq c-2)$.

**Step 4:** Calculate the $\chi^2$ test statistic and compare it with the $\chi^2_{c-1-1}$ distribution. The `stats.chisquare` function has an argument `ddof`, set to 0 by default, that is the amount by which you must diminish the degrees of freedom of the no-estimates chi-squared test that would use $c-1$ degrees of freedom. In this case, `ddof = 1`.
