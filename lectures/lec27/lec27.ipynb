## Randomized Controlled Trials

### Potential Outcomes
As in Data 8, the analysis starts by imagining that each unit $i$ has a potential treatment outcome $t_i$ and a potential control outcome $c_i$, before the experiment begins, and regardless of the group to which unit $i$ gets randomly assigned. We can think of the potential outcomes as the pairs $(t_i, c_i)$ for $i = 1, 2, \ldots, N$.

Note that there is no randomness in these lists of potential outcomes. They are just lists of numbers. The randomness will come in later, through the random assignment of units to the treatment and control groups. You can of course model the potential outcomes as random variables but we'll leave that for another course.

Let's set up some notation related to the potential outcomes.
- For the potential treatment outcomes, let the mean be $\bar{t} = \displaystyle \frac{1}{N} \sum_{i=1}^N t_i$, let the variance be $\sigma_t^2 = \displaystyle \frac{1}{N} \sum_{i=1}^N (t_i - \bar{t})^2$, and let $S_t^2 = \displaystyle \frac{N}{N-1}\sigma_t^2 = \frac{1}{N-1} \sum_{i=1}^N (t_i - \bar{t})^2$.
- Analogusly define $\bar{c}$, $\sigma_c^2$, and $S_c^2$.
- Let the covariance of the pairs $(t_i, c_i)$ be $cov(t, c) = \displaystyle \frac{1}{n} \sum_{i=1}^n (t_i - \bar{t})(c_i - \bar{c})$.

### Average Treatment Effect (ATE)
The main parameter of interest to us will be the *average treatment effect* or ATE, defined as
$$
\tau ~ = ~ \frac{1}{N} \sum_{i=1}^N (t_i - c_i) ~ = ~ \bar{t} - \bar{c}
$$

### The Randomization and the Data
In a randomized controlled trial, units are assigned at random to the treatment group or the control group. Crucially, no unit can be assigned to both groups, so we cannot observe both $t_i$ and $c_i$ for the same $i$.

We will assume a particular randomization, which is that the treatment group is a simple random sample of $n$ indices from among $\{1, 2, \ldots, N\}$. We will call this set of sampled indices $T$. The control group will be a simple random sample of size $m$ from the *remaining* indices $\{1, 2, \ldots, N\} \backslash T$. We will call this set of sampled indices $C$. By construction, $T$ and $C$ are disjoint.

Often, $C$ is equal to all the remaining indices $\{1, 2, \ldots, N\} \backslash T$ and hence $m = N-n$. That is, each of the $N$ units gets assigned to one group or the other. But you don't have to assume this in general.

Let $X_1, X_2, \ldots, X_n$ be the $t$-values of the indices in $T$. These are the observed treatment values. Let $Y_1, Y_2, \ldots, Y_m$ be the $c$-values of the indices in $C$. These are the observed control values.

Recall that $\bar{X}$ is an unbiased estimator of $\bar{t}$ and $\bar{Y}$ is an unbiased estimator of $\bar{c}$.

Also, $\sigma_X^2$ and $S_X^2$ can both be used as estimators of $\sigma_t^2$, and both $\sigma_Y^2$ and $S_Y^2$ are estimators of $\sigma_c^2$.

But we cannot calculate or estimate $cov(t, c)$ because we can't observe both $t_i$ and $c_i$ for the same index $i$.

## Testing Whether the Treatment has an Effect
We'd like to know if the treatment does anything. One way to approach this is to run a test of hypotheses.

### Null hypotheses
Informally, the null hypothesis is that the treatment does nothing. There are multiple ways of making this informal statement precise.

**Fisher's strong (or sharp) null** says that $t_i = c_i$ for each $i$. That is, for each unit, the potential outcome is the same regardless of the group to which they are assigned. Then $t_i - c_i = 0$ for each $i$ and $\tau = \bar{t} - \bar{c} = 0$.

**Neyman's weak null** assumes only that $\bar{t} = \bar{c}$, that is, the potential outcomes are equal on average. This allows the possibility that some units might benefit from the treatment while others might do worse. We will work under this weak null, which can be written as $\tau = 0$.

Here are two tests that you have seen before.

### Binary Outcomes: An Exact Test
In the case of binary outcomes, Neyman's weak null is equivalent to saying that the proportion of 1s among the potential treatment outcomes is the same as the proportion of 1s among the control outcomes, that is, the two sets of outcomes have the same Bernoulli distribution.

Under this hypothesis, it doesn't matter which unit is labeled "treatment" and which "control" as long as the two groups have sizes $n$ and $m = N-n$ respectively. We are assuming that each of the $N$ units is assigned to a group. 

The treatment group is a simple random sample of the $N$ units. If there are $w$ 1s in the pooled sample of $N$ units, then under null hypothesis of equal distributions, the distribution of the number of 1s in the treatment group is hypergeometric $(N, w, n)$. In this setting, *Fisher's exact test* uses the number of 1s in the treatment group as the test statistic, and its hypergeometric null distribution to find the $p$-value. See [Lecture 12](https://static.us.edusercontent.com/files/cHf8G5925uu3X0Mfqn99NDFV) Example 3 on page 11.

### Permutation Test
Under Fisher's strong null, or under Neyman's weak null applied to binary data, the distribution of all potential treatment outcomes is equal to the distribution of all potential control outcomes. In these cases we can do what was done in [Data 8](https://inferentialthinking.com/chapters/12/2/causality/#permutation-test): a permutation test. Use the difference of the observed means as the test statistic. To simulate its distribution under the null, do the following repeatedly: pool the sample and split it randomly into $n$ "treatment" values and $m$ "control" values, and find the difference of the two group means. Use the simulated null distribution to find the $p$-value.

## Estimating the Average Treatment Effect
Instead of testing hypotheses about the ATE $\tau = \bar{t} - \bar{c}$ it is more useful to be able to estimate it. We'll know roughly how big it is, not just whether or not it's zero.

### Unbiased Estimator
The parameter $\tau$ is the difference between the means of all the potential treatment outcomes and the potential control outcomes. The natural estimator $\hat{\tau}$ is the difference between the means of the observed treatment outcomes and the observed control outcomes.
$$
\hat{\tau} = \bar{X} - \bar{Y} = \frac{1}{n} \sum_{i \in T} t_i - \frac{1}{m} \sum_{i \in C} c_i
$$
As we observed earlier, the average of the observed treatment outcomes is an unbiased estimator of $\bar{t}$, and the average of the observed control outocmes is an unbiased estimator of $\bar{c}$. So the $\hat{\tau}$ is an unbiased estimator of the ATE $\bar{t} - \bar{c}$. 

### Variance of the Estimator
Recall the [variance of a simple random sample sum](https://data140.org/textbook/content/chapter-13/symmetry-and-indicators/#variance-of-a-simple-random-sample-sum). It's the same as the variance of the sum when the sampling is with replacement, times a correction factor. We now have
$$
\begin{align*}
Var(\hat{\tau}) &= Var(\bar{X}) + Var(\bar{Y}) -2Cov(\bar{X}, \bar{Y}) \\
&= \frac{1}{n^2} \cdot n\sigma_t^2 \frac{N-n}{N-1} + \frac{1}{m^2} \cdot m\sigma_c^2 \frac{N-m}{N-1} - 2Cov(\bar{X} - \bar{Y}) \\
&= \frac{\sigma_t^2}{n} \frac{N-n}{N-1} + \frac{\sigma_c^2}{m} \frac{N-m}{N-1} - 2Cov(\bar{X} - \bar{Y}) \\
&= \frac{N}{N-1}\left( \frac{\sigma_t^2}{n} + \frac{\sigma_c^2}{m} \right) - \frac{1}{N-1}(\sigma_t^2 + \sigma_c^2) - 2Cov(\bar{X} - \bar{Y})
\end{align*}
$$

In a worksheet exercise, you will show that $Cov(\bar{X}, \bar{Y}) = \displaystyle -\frac{1}{N-1}cov(t, c)$. It follows that
$$
\begin{align*}
Var(\hat{\tau}) &= \frac{N}{N-1}\left( \frac{\sigma_t^2}{n} + \frac{\sigma_c^2}{m} \right) + \frac{1}{N-1}(2cov(t, c) - \sigma_t^2 - \sigma_c^2) \\
&= \frac{S_t^2}{n} + \frac{S_c^2}{m} + \frac{1}{N-1}(2cov(t, c) - \sigma_t^2 - \sigma_c^2) \\
&\le \frac{S_t^2}{n} + \frac{S_c^2}{m}
\end{align*}
$$
Surprisingly and conveniently, this bound is approximately the variance of the difference between the means if the observed outcomes had been drawn independently of each other and with replacement from the potential treatment and control outcomes. If $n$ and $m$ are small relative to $N$ then all this independence will be approximately true. Regardless of the relative sizes of the samples and the population, the bound is pretty good for large $N$ because then $1/(N-1)$ is small. 

There are several ways of seeing why the bound works. For example, $\sigma_t^2 + \sigma_c^2 - 2cov(t, c)$ is the variance of the list $t_i - c_i$, $1 \le i \le N$, and hence non-negative. Alternatively $cov(t, c) \le \sigma_t\sigma_c$ by the correlation inequality and so $\sigma_t^2 + \sigma_c^2 - 2cov(t, c) \ge (\sigma_t - \sigma_c)^2 \ge 0$.

Though the line above the bound gives a formula for $Var(\hat{\tau})$, the formula involves $cov(t, c)$ which, as we observed earlier, can't be calculated or estimated. The bound can be estimated by replacing the population variances by the variances observed in the samples: $\displaystyle \frac{S_X^2}{n} + \frac{S_Y^2}{m}$.

The estimated upper bound can be used as a conservative estimate of $Var(\hat{\tau})$. Under some regularity conditions, $\hat{\tau}$ is roughly normal when $n$ and $m$ are large. In that case
$$
\hat{\tau} \pm 2 \sqrt{\frac{S_X^2}{n} + \frac{S_Y^2}{m} }
$$ 
is a conservative 95\% confidence interval for the ATE $\tau$.

#### Note: The Sign of $Cov(\bar{X}, \bar{Y})$
It is reasonable to suppose that $cov(t, c) > 0$, or in other words, the scatter plot of the points $(t_i, c_i)$ shows a postive association. In that case $Cov(\bar{X}, \bar{Y})$ will be negative. If many of the high $t$-values got picked for the treatment group, their unobservable $c$-values would also tend to be high, which implies that the observable $c$-values of the remaining group would tend to be low.

## The Standard Notation and Some Assumptions
Causal inference literature uses a fairly standard notation (some of it proposed by Neyman as early as 1923) for the fundamental building blocks of the subject. The notation in our discussion above has been lighter and a bit easier to work with. To go further, however, it will help to use the standard notation and formalization of the assumptions.

Consider a randomized experiment with a finite number of units. For each unit $i$, define the following notation.

### The main ingredients
- There are $n$ units in all, of which $n_1$ are in Group 1 and $n_0$ in Group 0. (These are $N$, $n$, and $m$ in our earlier notation.)
- $Z_i$ is the indicator of unit $i$ receiving the treatment.
- $Y_i(1)$ is the potential "treatment" outcome ($t_i$ above, though now it could be random) and $Y_i(0)$ is the potential "control" outcome ($c_i$ above).
- $\tau_i = Y_i(1) - Y_i (0)$ is the treatment effect for individual $i$.
- $\bar{\tau} = \displaystyle \frac{1}{n} \sum_{i=1}^n \tau_i$ is the average treatment effect, the main parameter in our discussions.

### Observed Value
$Y_i = Z_iY_i(1) + (1-Z_i)Y_i(0)$ is what we observe for unit $i$, that is, we observe $Y_i(1)$ if $Z_i = 1$ and we observe $Y_i(0)$ if $Z_i = 0$. There are multiple useful ways of describing $Y_i$. For example,
$$
Y_i ~ = ~ Y_i(0) + Z_i(Y_i(1) - Y_i(0)) ~ = ~ Y_i(0) + Z_i\tau_i
$$
Notice also that the average outcome of the treatment group is $\displaystyle \frac{1}{n} \sum_{i=1}^n Z_iY_i$ and the average outcome of the control group is $\displaystyle \frac{1}{n} \sum_{i=1}^n (1-Z_i)Y_i$. So the estimator of the average treatment effect can be written as
$$
\hat{\tau} ~ = ~ \displaystyle \frac{1}{n} \sum_{i=1}^n Z_iY_i - \frac{1}{n} \sum_{i=1}^n (1-Z_i)Y_i
$$

### Assumptions in a Randomized Experiment
We assume that the potential outcome pairs are independent across units. That is, the potential outcomes of unit $i$ are not affected by any other unit.

We also assume that $Z_i$ is independent of the pair $(Y_i(1), Y_i(0))$. One way to think about this is that the potential outcomes are properties of the unit and the treatment, whereas $Z_i$ comes from an independent mechanism for assigning units to the two groups.

Notation for independence: $\{Y_i(1), Y_i(0)\} \perp \!\!\! \perp Z_i $

## Observational Studies
In an observational study, the units "assign themselves" to the two groups. It is no longer reasonable to make the independence assumption above, because the potential outcomes as well as the assignment might both be influenced by the same confounding variable.

As an example, in the mid-20th century medicine had yet to establish that smoking causes lung cancer. At that time it was observed that coffee drinkers had higher rates of lung cancer than those who didn't drink coffee. The finger pointed at coffee as a cause of lung cancer. 

In our notation, $Y_i(1)$ and $Y_i(0)$ could be indicators of Person $i$ getting lung cancer, and $Z_i$ would be the indicator of Person $i$ being a coffee drinker.

But smoking at a cafe was a common activity, so smoking was a confounding factor. In our example, $X_i$ could be the indicator of Person $i$ being a smoker.  

### Conditional Independence Assumption
For each $i$, we will assume $\{Y_i(1), Y_i(0)\} \perp \!\!\! \perp Z_i \mid X_i$. That is, the potential outcomes are conditionally independent of the group assignment, given the confounder.

In our example, we are saying that among the smokers, whether or not the person is a coffee drinker is independent of their potential scores, and the same is true among the non-smokers. Roughly, in a simplified world where smoking is the only confounder in the study, we are assuming that in the population of smokers we have one of our "no confounder" experiments, and in the population of non-smokers we have another of our "no confounder" experiments.

We will also assume that the units are i.i.d. copies of each other. Some simplifying notation:
- $\{X_i, Z_i, Y_i(1), Y_i(0)\}_{i=1}^n \stackrel{iid}{\sim} \{X, Z, Y(1), Y(0)\}$ where the right hand side is a generic draw from the underlying joint distribution of each unit.
- In this notation, $\{Y(1), Y(0)\} \perp \!\!\! \perp Z \mid X$

### Average Treatment Effect
Before we get to the familiar average treatment effect, let us define the *conditional average treatment effect* (or CATE) to be
$$
\tau(X) ~ = ~ E\left( (Y(1) - Y(0)) \mid X \right)
$$
If the confounder is binary as in our smoking example, there are just two values of $\tau(X)$. One is the expected treatment effect among the smokers, and the other is the expected treatment effect among the non-smokers.

By iteration, the average treatment effect is
$$
\tau = E\left[\tau(X)\right]
$$
### Propensity Score
Our familiar estimator of the average treatment effect is
$$
\hat{\tau} ~ = ~ \frac{1}{n} \sum_{i=1}^n Z_iY_i - \frac{1}{n} \sum_{i=1}^n (1-Z_i)Y_i
$$
But in an observational study, this estimator could be biased. For example, in the example of coffee and lung cancer, smokers are more likely to be in the coffee group and more likely to get cancer, leading to an overestimate.

To bring more balance to the contributions of the two groups, we can try to measure the tendency, or *propensity,* of smokers to be coffee drinkers. We define the *propensity score* 
$$
e(X, Y(1), Y(0)) ~ = ~ P(Z=1 \mid X, Y(1), Y(0)) ~ = ~ P(Z=1 \mid X)
$$
by our conditional independence assumption. So the propensity score can be written as
$$
e(X) ~ = ~ P(Z = 1 \mid X) ~ = ~ E(Z \mid X)
$$
In our example, $e(X)$ has two values: the probability of being a coffee drinker given that the person is a smoker, and the probability of being a coffee drinker given that the person is a non-smoker. These can be estimated respectively by the proportion of coffee drinkers among the smokers and the proportion of coffee drinkers among the non-smokers. These estimated proportions, or any other estimator of $e(X)$, will be denoted $\hat{e}(X)$.

### Inverse Propensity Score Weighting
We expect the proportion of coffee drinkers among smokers to be high. Since the terms in the average treatment outcome will be too high, we can dampen their effect by weighting them by the *inverse of the propensity score*, and placing correspondingly higher weights on the terms in the average control outcome. Define the *inverse propensity score weighted estimator* as
$$
\tau_{IPW} = \frac{1}{n} \sum_{i=1}^n \frac{Z_iY_i}{\hat{e}(X_i)} - \frac{1}{n} \sum_{i=1}^n \frac{(1-Z_i)Y_i}{1-\hat{e}(X_i)}
$$
It turns out that this provides the right balance. If the propensity scores are known and don't have to be estimated, the estimator above would use $e$ instead of $\hat{e}$, and it would be unbiased under additional assumptions including assuming that there are no unmeasured confounders. 

Usually, the propensity scores have to be estimated. In that case, under certain conditions, the estimator above may be asymptotically unbiased depending on how the propensity scores are estimated.

This is where the lecture ended. For those who are curious, here's a result that motivates the estimator above.

### Optional: Motivating $\tau_{IPW}$
We will show that if $\{Y(1), Y(0)\} \perp \!\!\! \perp Z \mid X$ and $0 < e(X) < 1$, then
$$
E(Y(1)) ~ = E\left[ \frac{ZY}{e(X)} \right], ~~~~~~~~~~ E(Y(0)) ~ = E\left[ \frac{(1-Z)Y}{1-e(X)} \right]
$$
and hence
$$
\tau ~ = ~ E\left[ Y(1) - Y(0) \right] ~ = ~ E\left[\frac{ZY}{e(X)} - \frac{(1-Z)Y}{1-e(X)} \right]
$$
We can therefore think of $\hat{\tau}_{IPW}$ as a method of moments estimator of $\tau$.

We will show $E(Y(1)) ~ = \displaystyle E\left[ \frac{ZY}{e(X)} \right]$. The second result is proved analogously and the expression for $\tau$ follows.

Note the condition $0 < e(X) < 1$. If any propensity score is either 0 or 1, $\hat{\tau}_{IPW}$ will blow up.

$$
\begin{align*}
E\left[ \frac{ZY}{e(X)} \right] &= E\left[ \frac{ZY(1)}{e(X)} \right] \\
&= E\left[ E\left[ \frac{ZY(1)}{e(X)} \mid X \right] \right] \\
&= E\left[ \frac{1}{e(X)}E\left[ZY(1) \mid X \right] \right] \\
&= E\left[ \frac{1}{e(X)} E\left[Z \mid X \right]E\left[Y(1) \mid X \right] \right] ~~ \text{by conditional independence}\\ 
&= E\left(E\left[Y(1) \mid X \right]\right) ~~ \text{by the definition of } e(X) \\
&= E\left(Y(1)\right)
\end{align*}
$$
