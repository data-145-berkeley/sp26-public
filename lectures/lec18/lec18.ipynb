## One Sample of Numerical Data: Tests for the Mean

Let $X_1, X_2, \ldots, X_n$ be i.i.d. with mean $\mu$ and variance $\sigma^2$. Suppose you want to test the null hypothesis $H_0: \mu = \mu_0$. It's worth thinking about when you might want to do that. Examples:
- The case $H_0: \mu = 0$ comes up often, as we will see below.
- If there is an external standard like "the normal temperature of the human body is 98.6 degrees Fahrenheit," you might want to test $H_0: \mu = 98.6$.
- The case where $X_i$ is Bernoulli $(p)$ comes up often. You might want to test $H_0: p = 1/2$, for example.

#### Test Statistic
The basic statistic is $\displaystyle T_n = \frac{\bar{X}_n - \mu_0}{\sigma/\sqrt{n}}$.

#### Null distribution
Now start looking at the sample size and whether or not you know $\sigma$.
1. **Large sample:** If $n$ is large, the CLT says the null distribution of $T_n$ is approximately standard normal. By an earlier exercise, you can replace an unknown $\sigma^2$ with either its MLE $\hat{\sigma}^2$ or its unbiased estimator $S^2$, and the standard normal approximation still holds. Note that in the Bernoulli setting, fixing $p_0$ determines not only the expectation but also the variance of $T_n$, because $Var(T_n) = \displaystyle \sqrt{\frac{p_0(1-p_0)}{n}}$.
2. **Small sample:** Now you're going to need some help. For example, you might want to make some assumptions about the underlying distribution of the sampled elements.
   - If you believe the underlying distribution is normal, and you know $\sigma^2$, then the null distribution of $T_n$ is standard normal for each $n$.
   - If you believe the underlying distribution is normal, and you don't know $\sigma^2$, then replace $\sigma$ by $S$ to get the statistic $\displaystyle T_n^* = \frac{\bar{X}_n - \mu_0}{S/\sqrt{n}}$. The null distribution of this statistic is $t$ with $n-1$ degrees of freedom for each $n > 1$. See the Section 4 on nuisance parameters in Lecture 13.
   - If you are testing for $p$ in a Bernoulli setting, use $\sum_i X_i$ as the statistic. Its null distribution is binomial $(n, p_0)$.

### Special Case: Paired Data

Sometimes what seems like two samples is actually one sample of pairs.

For example, suppose our data consist of two observations (say two test scores) on each of $n$ individuals, and we want to see if there is any difference between the distributions of the two scores. In this case, even though it might be reasonable to assume that individuals are independent of each other, the two scores of a single individual's scores are not independent of each other. Therefore, even though we have two samples of scores, the two samples are *paired*. Their is one pair of dependent scores for each individual.

Formally, the data are $(X_1, Y_1), (X_2, Y_2), \ldots, (X_n, Y_n)$. The $n$ pairs $(X_i, Y_i)$ are independent of each other and identically distributed. But for each fixed $i$, the random variables $X_i$ and $Y_i$ are dependent.

If you think of the $X$s as baseline measurements on patients, and the $Y$s as measurements after a medical treatment, then it is natural to ask whether the treatment had any effect. To study this, it is also natural to consider the i.i.d. differences $D_i = X_i - Y_i$ for $1 \le i \le n$, thus **reducing the problem to one sample of differences**.

Our null hypothesis is that the treatment doesn't do anything. We will formulate it as follows:

$H_0: (X, Y) \stackrel{d}{=} (Y, X)$

That is, the joint distribution of $X$ and $Y$ is symmetric. It doesn't matter which variable you call $X$ and which you call $Y$. 

For simplicity, assume that $X$ and $Y$ have a joint density surface so that the differences can't be $0$ no two differences can be equal. Otherwise our calculations will have to account for ties of various kinds. The ties can be dealt with but we won't study that case.

For any $x > 0$, $P(X-Y > x) = P(Y-X < -x) = P(X-Y < -x)$ under $H_0$.

Thus the distribution of $D = X-Y$ is symmetric about $0$ under $H_0$. You will sometimes see weaker versions of this hypothesis, such as "the median difference is $0$."

So you could test $H_0: \mu_D = 0$ by using everything you know about one-sample tests for the mean.

#### A Nonparametric method: Sign Test
Under the assumption that there are no ties, each $D_i$ is either positive or negative. It can't be $0$. Then under $H_0$, each $D_i$ is positive with chance $1/2$ independent of all other differences. 

That is, under $H_0$ the number of positive differences has the binomial $(n, 1/2)$ distribution.

So you can use the number of positive differences as the test statistic, and compare it to the binomial $(n, 1/2)$ distribution.

This is easy and works for all sample sizes. But it throws away the data apart from the signs of the differences.

## Two Independent Numerical Samples: Equality of Means
Suppose we have two **independent** samples $X_1, X_2, ..., X_n$ and $Y_1, Y_2,..., Y_m$, both i.i.d. The null hypothesis is that these two samples are drawn from two underlying populations with the same mean (not necessarily the same distribution). 

$H_0: \mu_X = \mu_Y$

A natural test statistic to consider is $\bar{X} - \bar{Y}$, the difference between the sample means. Depending on our alternative hypothesis, we may need to make slight modifications (e.g. taking absolute value), but at the heart of it is the difference in sample means. 

Regardless of the underlying distributions of $X$ and $Y$, the mean and variance of our statistic are:
$$E(\bar{X} - \bar{Y}) = \mu_{X} - \mu_{Y}, ~~~~~
Var(\bar{X} - \bar{Y}) = \frac{\sigma_{X}^{2}}{n} + \frac{\sigma_{Y}^{2}}{m}$$

#### Test Statistic
Under $H_0$, we want to measure how far our statistic is from 0. Construct a new statistic $T_n$ that measures this in standard units.
$$ T = \frac{(\bar{X} - \bar{Y}) - 0}{\sqrt{\frac{\sigma_X^2}{n} + \frac{\sigma_Y^2}{m}}} $$

#### Null Distribution
Just as in the one-sample case, we have to consider our sample size and whether or not we know $\sigma_X$ and $\sigma_Y$.

1. **Large samples:** If $n$ and $m$ are large, the CLT says the distribution of 
$$ 
\frac{(\bar{X} - \bar{Y}) - 0}{\sqrt{\frac{\mathord{?}_{X}^{2}}{n} + \frac{\mathord{?}_{Y}^{2}}{m}}} 
$$ 
is approximately standard normal. Here $\mathord{?}$ can be either $\sigma^{2}$ or $\hat{\sigma}^{2}$ or $S^{2}$.

2. **Small samples:** As before, you're going to need some help, perhaps in the form of assumptions about the underlying distribution.
- If you believe that both the samples are normal, and $\sigma_{X}$ and $\sigma_{Y}$ are known, then the null distribution of $T$ is standard normal for all sample sizes $n > 1$ and $m > 1$.

- If both the samples are normal, and $\sigma_{X}$ and $\sigma_{Y}$ are *unknown*, then switch to the statistic
$$ 
T_2 = \frac{(\bar{X} - \bar{Y}) - 0}{\sqrt{\frac{S_{X}^{2}}{n} + \frac{S_{Y}^{2}}{m}}}
$$
The null distribution of this statistic is [complicated](https://en.wikipedia.org/wiki/Welch%27s_t-test). We won't touch it. The good news is that a conservative estimate of the $p$-value can be found by using the $t$-distribution with degrees of freedom equal to the smaller of $n-1$ and $m-1$. We won't prove that either.

- If you are in the strange situation where both the samples are normal, with unknown *but equal* variances, then the statistic above becomes
$$ 
T_{3} ~=~ \frac{(\bar{X} - \bar{Y}) - 0}{\sqrt{\frac{S_p^2}{n} + \frac{S_p^2}{m}}} ~=~ 
\frac{(\bar{X} - \bar{Y}) - 0}{S_p\sqrt{\frac{1}{n} + \frac{1}{m}} }
$$
where 
$$ S_p^{2} = \frac{(n-1)S_{X}^{2} + (m-1)S_{Y}^{2}}{n+m-2} $$
is called the *pooled sample variance* and is the common estimate of population variance $\sigma^{2}$ for both samples.

    For all sample sizes, the distribution of $T_3$ is $t$ with $n+m-2$ degrees of freedom. For a derivation, **see the two-sample $t$ test in Lecture 13**.

## Two Independent Numerical Samples: Equality of Distributions

Suppose $X_1, X_2,..., X_n$ are iid from a distribution with cdf $F_X$. Independent of $X$, suppose $Y_1, Y_2, ..., Y_m$ are iid from a distribution with cdf $F_Y$. Assume both $F_X$ and $F_Y$ are continuous and increasing. 

Our null hypothesis is that the two samples are drawn from the same distribution. In other words, $H_0$ says $F_X = F_Y$.

#### Nonparametric Method: Two-sample K-S Test
**See Lecture 11.** This is the third K-S test, where the statistic is the K-S distance between the two empirical cdfs. The null distribution of the statistic is obtained by simulation:
- Shuffle the pooled sample.
- Split it into two sets of $n$ and $m$ values; the first $n$ and the last $m$ will do.
- Calcualate the K-S statistic.
- Repeat the three steps above numerous times. But it's usually not feasible to generate all $(n+m)!$ possible permutations.

#### Permutation preserves the level
The "new samples" generated by permutation can't have values that weren't in the original samples. But of course there are values outside the original samples. We're not using them? Does this cause problems? 

It's worth looking at the level informally. Our simulated null distribution is a conditional distribution given the original pooled sample which we will call $X_p$. If we specify the level to be $\alpha$, then we need $P_{H_0}(\text{reject } H_0)$ to be $\alpha$. Our testing algorithm ensures (assuming that the underlying distributions are well-behaved) that conditionally given $X$ our test has level $\alpha$. So
$$
P_{H_0}(\text{reject } H_0) = E_{H_0}(I(\text{reject } H_0)) = E_{H_0}(E(I(\text{reject } H_0) \mid X_p) = E_{H_0}(\alpha) = \alpha
$$

#### Another Nonparametric Method: The Wilcoxon Rank Sum Test

Sometimes we want to see if one distribution is generally more likely to produce smaller values than the other. You saw this in the exercise comparing the birth weights of babies of smokers and non-smokers. One idea for how to work under the null hypothesis of equal distributions is to replace each value in the pooled sample by its rank in the pooled sample. 

Here is an example of two small samples and their ranks in the pooled sample. Rank 1 is the smallest value.
|  |$x_1$|$x_2$|$x_3$|$y_1$|$y_2$|
|--:|:---:|:---:|:---:|:---:|:---:|
**value** | 2 | 4 | 1 | -100 | 3 |
**rank**  | 3 | 5 | 2 | 1   | 4 |

We will assume that our $F_X$ and $F_Y$ are continuous so we don't have to deal with ties. This test does work without this assumption, but it's messy. 

Under the null hypothesis that $F_X = F_Y$, the ranks of the pooled sample $X_1, X_2, \ldots, X_n, Y_1, Y_2, \ldots, Y_m$ are a random permutation of all the ranks $1, 2, \ldots, N$ where $N = n+m$.

#### Test Statistic: Wilcoxon Rank Sum
Let the statistic $W$ be the sum of the ranks of the $X$-sample. If $W$ is very low, then the $X$-sample is at the low end of the pooled sample instead of being sprinkled uniformly as the null hypothesis would predict. So we would want to reject the null. 

#### Null Distribution
To understand the null distribution better, review the calculations of the [expectation and variance of the sum of a simple random sample](https://data140.org/textbook/content/chapter-13/symmetry-and-indicators/#variance-of-a-simple-random-sample-sum).

In our case, the population is $1, 2, \ldots, N$, and we have a simple random sample of size $n$. By symmetry in simple random sampling, each sampled element is uniform on $1, 2, \ldots, N$. Hence each sampled element has expectation $\displaystyle \frac{N+1}{2}$ and variance $\displaystyle \frac{N^2 - 1}{12}$ by facts about the uniform distribution on consecutive integers. It follows that
$$
E(W) = n \cdot \frac{N+1}{2} = \frac{n(n+m+1)}{2}
$$
and
$$
Var(W) = n \cdot \frac{N^2 - 1}{12} \cdot \frac{N-n}{N-1} = n \cdot \frac{N+1}{12} \cdot (N-n) = \frac{nm(n+m+1)}{12}
$$

- Chebyshev's bound will give you a rough range where the rank sum statistic is overwhelmingly likely to fall.
- Or you can simulate the rank sum statistic under the null hypothesis and compare your observed rank sum to the simulated distribution. You will carry out this plan in exercises.
- It turns out that if $n$ and $m$ are large then the distribution of $W$ is approximately normal. In that case all inference can be done based on the normal curve, using the expectation and variance calculated above.
