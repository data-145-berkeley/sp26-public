## 1. The Canonical Linear Model in Full Generality

### 1.1 Recap from Tuesday

We ended Lecture 23 with an $n = 3$ warm-up: three independent observations $Z_0, Z_1, Z_2$, each $N(\cdot, \sigma^2)$, with

- $E[Z_0] = \mu_0$ unrestricted (nuisance),
- $E[Z_1] = \mu_1$ tested ($H_0: \mu_1 = 0$),
- $E[Z_2] = 0$ (pure noise).

The key insight: $Z_0$ is useless — its mean is free under both hypotheses and it's independent of everything else. Drop it; test on $(Z_1, Z_2)$ alone.

Now we extend this to arbitrary dimensions.

### 1.2 General Setup

Observe
$$Z \sim N_n(\mu,\ \sigma^2 I_n), \qquad \sigma^2 > 0 \text{ (known or unknown)}.$$

Partition $Z$ into three blocks of dimensions $d_0 + d_1 + d_r = n$:

- $Z_0 \in \mathbb R^{d_0}$: **nuisance** — $E[Z_0] = \mu_0$, unrestricted under both $H_0$ and $H_1$.
- $Z_1 \in \mathbb R^{d_1}$: **tested** — $H_0: \mu_1 = 0$ vs $H_1: \mu_1 \neq 0$.
- $Z_r \in \mathbb R^{d_r}$: **pure noise** — $E[Z_r] = 0$ under both hypotheses.

All three blocks are independent (the covariance matrix $\sigma^2 I_n$ is diagonal). Ignore $Z_0$. Under $H_0$, both $Z_1$ and $Z_r$ have mean zero and are iid $N(0, \sigma^2)$ in each coordinate.

### 1.3 The Four Tests

Under $H_0$: $Z_1 \sim N_{d_1}(\mathbf 0, \sigma^2 I_{d_1})$ and $Z_r \sim N_{d_r}(\mathbf 0, \sigma^2 I_{d_r})$, independent.

Define the variance estimate $\hat\sigma^2 = \|Z_r\|^2 / d_r$ (unbiased for $\sigma^2$ since $\|Z_r\|^2 \sim \sigma^2 \chi^2_{d_r}$).

| | $\sigma^2$ known | $\sigma^2$ unknown |
|---|---|---|
| **$d_1 = 1$** | $z$-test: $Z_1/\sigma \sim N(0, 1)$ | $t$-test: $Z_1/\hat\sigma \sim t_{d_r}$ |
| **$d_1 \geq 1$** | $\chi^2$-test: $\|Z_1\|^2/\sigma^2 \sim \chi^2_{d_1}$ | $F$-test: $\dfrac{\|Z_1\|^2/d_1}{\|Z_r\|^2/d_r} \sim F_{d_1,\, d_r}$ |

All four tests reject for large values of the statistic.

**One-sided tests** are available when $d_1 = 1$: use the signed $z$- or $t$-statistic. The $\chi^2$ and $F$ forms are inherently two-sided.

**Connecting to Lecture 23.** The one-sample $t$-test with $n$ observations is the case $(d_0, d_1, d_r) = (0, 1, n - 1)$: no nuisance, 1D signal, $(n-1)$-dimensional residual. The $n = 2$ canonical model was $(0, 1, 1)$, giving a $t_1$ (Cauchy) test. The $n = 3$ warm-up was $(1, 1, 1)$.

### 1.4 Confidence Intervals from the Canonical Model

Recall from Lecture 14: a confidence interval is the set of parameter values not rejected by the corresponding test. We can invert any of the four tests above to get a CI or confidence region.

**The most common case: $d_1 = 1$, $\sigma^2$ unknown.** To test $H_0: \mu_1 = \mu_1^0$ (for any specific value $\mu_1^0$), we shift: $(Z_1 - \mu_1^0)/\hat\sigma \sim t_{d_r}$ under $H_0$. We fail to reject when $|Z_1 - \mu_1^0| \leq \hat\sigma \cdot t_{d_r}(\alpha/2)$. Inverting:

$$\mu_1 \in Z_1 \pm \hat\sigma \cdot t_{d_r}(\alpha/2).$$

**$d_1 = 1$, $\sigma^2$ known:** $\mu_1 \in Z_1 \pm \sigma \cdot z_{\alpha/2}$ — the familiar normal CI.

We'll see these again in §4 when we do regression: $Z_1$ becomes $\hat\beta_j$, $\hat\sigma$ becomes $\widehat{\text{SE}}(\hat\beta_j)$, and the interval becomes $\hat\beta_j \pm t_{n-d}(\alpha/2) \cdot \widehat{\text{SE}}(\hat\beta_j)$.

## 2. The General Linear Model Framework

### 2.1 From Canonical to General

In real problems we don't observe data in canonical coordinates. Instead we observe
$$Y \sim N_n(\theta,\ \sigma^2 I_n),$$
where the mean vector $\theta$ is constrained to lie in a **model subspace** $\Theta \subseteq \mathbb R^n$ (a linear subspace).

We want to test
$$H_0: \theta \in \Theta_0 \quad \text{vs} \quad H_1: \theta \in \Theta \setminus \Theta_0,$$
where $\Theta_0 \subseteq \Theta$ are nested subspaces with $\dim(\Theta_0) = d_0$ and $\dim(\Theta) = d_0 + d_1$.

The residual degrees of freedom are $d_r = n - d_0 - d_1 = n - \dim(\Theta)$.

### 2.2 The Rotation Recipe

The idea: build an orthogonal matrix that rotates $Y$ into canonical form.

**Step 1.** Choose orthonormal bases for three mutually orthogonal subspaces:

- $Q_0$ ($n \times d_0$): orthonormal basis for $\Theta_0$ (the null model space).
- $Q_1$ ($n \times d_1$): orthonormal basis for $\Theta \cap \Theta_0^\perp$ (the "extra directions" in the full model, orthogonal to the null model).
- $Q_r$ ($n \times d_r$): orthonormal basis for $\Theta^\perp$ (directions entirely outside the model).

Stack them: $Q = [Q_0\ |\ Q_1\ |\ Q_r]$ is an $n \times n$ orthogonal matrix.

**Step 2.** Rotate: $Z = Q^T Y$. By the affine transformation formula,
$$Z \sim N_n(Q^T \theta,\ \sigma^2 I_n).$$

**Step 3.** Read off the canonical structure.

- Under $H_0$ ($\theta \in \Theta_0$): $Q_1^T \theta = 0$ (since $\theta \perp$ the $Q_1$ directions) and $Q_r^T \theta = 0$ (since $\theta \in \Theta$, hence $\perp \Theta^\perp$). The nuisance block $Q_0^T \theta$ is free.
- Under $H_1$ ($\theta \in \Theta \setminus \Theta_0$): $Q_1^T \theta \neq 0$ (the component in the "extra" directions is nonzero), while $Q_r^T \theta = 0$ still.

**This is exactly the canonical model from §1.** Apply the appropriate test from the four-case table.

The rotation doesn't change the data or the geometry — only our coordinate system. The projections, residuals, and test statistics are all intrinsic (they don't depend on the choice of basis within each subspace).

## 3. Two-Sample $t$-Test: A Detailed Rotation

### 3.1 Setup

Two groups of normal observations with a common (unknown) variance:

- Group 1: $X_1, \ldots, X_m \stackrel{iid}{\sim} N(\mu, \sigma^2)$.
- Group 2: $Y_1, \ldots, Y_n \stackrel{iid}{\sim} N(\nu, \sigma^2)$.

Total sample size $N = m + n$. Stack everything into one vector:
$$W = (X_1, \ldots, X_m,\ Y_1, \ldots, Y_n)^T \in \mathbb R^N.$$

Then $W \sim N_N(\theta, \sigma^2 I_N)$ where $\theta = (\mu, \ldots, \mu,\ \nu, \ldots, \nu)^T$ (first $m$ entries equal to $\mu$, last $n$ entries equal to $\nu$).

Test $H_0: \mu = \nu$ vs $H_1: \mu \neq \nu$.

### 3.2 Identify the Subspaces

**Full model space** $\Theta$ (under $H_1$). The mean vector has the form $\theta = \mu \mathbf a + \nu \mathbf b$ where
$$\mathbf a = (\underbrace{1, \ldots, 1}_m,\ \underbrace{0, \ldots, 0}_n)^T, \qquad \mathbf b = (\underbrace{0, \ldots, 0}_m,\ \underbrace{1, \ldots, 1}_n)^T.$$
So $\Theta = \text{Span}(\mathbf a, \mathbf b)$, dimension $d_0 + d_1 = 2$.

**Null model space** $\Theta_0$ (under $H_0$). If $\mu = \nu$, then $\theta$ is a constant vector in $\mathbb R^N$. So $\Theta_0 = \text{Span}(\mathbf 1_N)$, dimension $d_0 = 1$.

**Dimensions:** $d_0 = 1$, $d_1 = 1$, $d_r = N - 2$.

### 3.3 Build the Orthonormal Basis

**$Q_0$ direction** (null model, dimension 1). The unit vector in $\Theta_0$:
$$\mathbf q_0 = \frac{\mathbf 1_N}{\sqrt N}.$$
This projects onto the grand mean. Ignoring $\mathbf q_0^T W$ means nothing changes if we add the same value to *every* observation $X_1, \ldots, X_m$ and $Y_1, \ldots, Y_n$: it won't affect the observed difference in means or our estimate of $\sigma^2$.

**$Q_1$ direction** (signal, dimension 1). We need a unit vector in $\Theta$ orthogonal to $\Theta_0$. Start with $\mathbf a$ and subtract its projection onto $\mathbf 1_N$:
$$\mathbf a - \frac{\mathbf a^T \mathbf 1_N}{N} \mathbf 1_N = \mathbf a - \frac{m}{N}\mathbf 1_N = \left(\underbrace{\frac{n}{N}, \ldots, \frac{n}{N}}_m,\ \underbrace{-\frac{m}{N}, \ldots, -\frac{m}{N}}_n\right)^T.$$

This vector is orthogonal to $\mathbf 1_N$ by construction. But rather than normalize it directly, it's clearer to first rescale it to a **contrast vector** with a natural interpretation. Multiply by $N/(mn)$:
$$\mathbf c = \frac{N}{mn}\left(\mathbf a - \frac{m}{N}\mathbf 1_N\right) = \left(\underbrace{\frac{1}{m}, \ldots, \frac{1}{m}}_m,\ \underbrace{-\frac{1}{n}, \ldots, -\frac{1}{n}}_n\right)^T.$$

Now observe:
$$\mathbf c^T W = \frac{1}{m}\sum_{i=1}^m X_i - \frac{1}{n}\sum_{j=1}^n Y_j = \bar X - \bar Y.$$

The contrast vector literally picks out the **difference of group means** — exactly the quantity we want to test. Normalize:
$$\|\mathbf c\|^2 = m \cdot \frac{1}{m^2} + n \cdot \frac{1}{n^2} = \frac{1}{m} + \frac{1}{n}, \qquad \mathbf q_1 = \frac{\mathbf c}{\sqrt{1/m + 1/n}}.$$

**$Q_r$** (residual, dimension $N - 2$). Any orthonormal basis for $\Theta^\perp$ — the $(N-2)$-dimensional subspace orthogonal to both $\mathbf a$ and $\mathbf b$. We won't need the explicit basis; only $\|Z_r\|^2$ matters.

### 3.4 Read Off the Test

**Signal coordinate.** Using $\mathbf c^T W = \bar X - \bar Y$:
$$Z_1 = \mathbf q_1^T W = \frac{\mathbf c^T W}{\sqrt{1/m + 1/n}} = \frac{\bar X - \bar Y}{\sqrt{1/m + 1/n}}.$$

Under $H_0$: $Z_1 \sim N(0, \sigma^2)$ (the $\sqrt{1/m + 1/n}$ in the denominator is exactly what standardizes $\bar X - \bar Y$ to have variance $\sigma^2$).

**Residual magnitude.** Using the orthogonal decomposition $\|W\|^2 = Z_0^2 + Z_1^2 + \|Z_r\|^2$:
$$\|Z_r\|^2 = \|W\|^2 - Z_0^2 - Z_1^2.$$

A calculation (expand and simplify) gives
$$\|Z_r\|^2 = \sum_{i=1}^m (X_i - \bar X)^2 + \sum_{j=1}^n (Y_j - \bar Y)^2 = (m - 1)S_X^2 + (n - 1)S_Y^2,$$
the **pooled residual sum of squares**, with $d_r = N - 2$ degrees of freedom.

**Pooled variance estimate:**
$$\hat\sigma^2 = S_p^2 = \frac{(m-1) S_X^2 + (n-1) S_Y^2}{N - 2}.$$

**$t$-statistic** ($d_1 = 1$, $\sigma^2$ unknown):
$$T = \frac{Z_1}{\hat\sigma} = \frac{\bar X - \bar Y}{S_p\sqrt{1/m + 1/n}} \sim t_{N - 2} \quad \text{under } H_0.$$

This is the classical **equal-variance two-sample $t$-test**.

**Confidence interval** for $\mu - \nu$ (by §1.4 inversion):
$$(\bar X - \bar Y) \pm t_{N-2}(\alpha/2) \cdot S_p \sqrt{\frac{1}{m} + \frac{1}{n}}.$$

## 4. Linear Regression

### 4.1 Setup

Observe
$$Y_i = x_i^T \beta + \epsilon_i, \qquad \epsilon_i \stackrel{iid}{\sim} N(0, \sigma^2), \qquad i = 1, \ldots, n.$$

Here $x_i \in \mathbb R^d$ is a fixed predictor vector, $\beta \in \mathbb R^d$ is the parameter, and $\epsilon_i$ is noise. The **design matrix** is $X \in \mathbb R^{n \times d}$ with rows $x_i^T$. Assume $X$ has full column rank ($d \leq n$).

In matrix form:
$$Y \sim N_n(X\beta,\ \sigma^2 I_n).$$

This is a special case of the general linear model: $\theta = X\beta$ is the mean vector, constrained to lie in $\text{Col}(X)$.

### 4.2 Model Space and Null Model Space

**Full model space.** $\Theta = \{X\beta : \beta \in \mathbb R^d\} = \text{Col}(X)$, dimension $d$.

The mean vector $\theta = E[Y]$ lives in $\text{Col}(X)$ — it's a linear combination of the columns of $X$, with coefficients $\beta$.

**Null hypothesis.** Partition $\beta = (\beta_1^T, \beta_{-1}^T)^T$ and correspondingly $X = [X_1\ |\ X_0]$, where $X_1$ are the columns being tested and $X_0$ the rest. Test

$$H_0: \beta_1 = 0, \qquad \text{i.e., } \theta \in \Theta_0 = \text{Col}(X_0).$$

**Dimensions:** $d_0 = \text{rank}(X_0)$ (number of "kept" predictors), $d_1 = d - d_0$ (number of "tested" predictors), $d_r = n - d$ (residual degrees of freedom).

### 4.3 OLS as Projection

Recall from Data 140 Chapter 25 that the **OLS estimator** is
$$\hat\beta_{\text{OLS}} = (X^T X)^{-1} X^T Y.$$

The **fitted values** are
$$\hat Y = X \hat\beta = X(X^T X)^{-1} X^T Y = P_\Theta Y,$$
where $P_\Theta = X(X^T X)^{-1} X^T$ is the **orthogonal projection matrix** onto $\text{Col}(X)$.

This is exactly our geometric framework: the OLS fit $\hat Y$ is the point in the model subspace $\Theta$ closest to $Y$. It's the orthogonal projection of $Y$ onto $\Theta$.

**Residuals.** The residual vector is
$$r = Y - \hat Y = (I - P_\Theta) Y = P_{\Theta^\perp} Y,$$
the projection of $Y$ onto the orthogonal complement of $\Theta$.

**Residual sum of squares.** $\text{RSS} = \|r\|^2 = \|P_{\Theta^\perp} Y\|^2 = \|Z_r\|^2$.

Since $Z_r \sim N_{d_r}(\mathbf 0, \sigma^2 I_{d_r})$, we have $\text{RSS}/\sigma^2 \sim \chi^2_{n-d}$, and
$$\hat\sigma^2 = \frac{\text{RSS}}{n - d}$$
is unbiased for $\sigma^2$. This is the familiar $S^2$ from the regression output.

### 4.4 The $F$-Test for a Subset of Coefficients

To test $H_0: \beta_1 = 0$ (a subset of coefficients), we compare two nested models:

- **Full model RSS:** $\text{RSS} = \|P_{\Theta^\perp} Y\|^2$ (fit with all $d$ predictors).
- **Null model RSS:** $\text{RSS}_0 = \|P_{\Theta_0^\perp} Y\|^2$ (fit with only the $d_0$ "kept" predictors).

Since $\Theta_0 \subseteq \Theta$, we have $\Theta^\perp \subseteq \Theta_0^\perp$, so $\text{RSS}_0 \geq \text{RSS}$. The difference
$$\text{RSS}_0 - \text{RSS} = \|Z_1\|^2$$
is the **signal magnitude** — how much the fit improves by adding the tested predictors.

The **$F$-statistic** is
$$F = \frac{(\text{RSS}_0 - \text{RSS})/d_1}{\text{RSS}/d_r} = \frac{\|Z_1\|^2 / d_1}{\|Z_r\|^2 / d_r} \sim F_{d_1,\, d_r} \quad \text{under } H_0,$$
where $d_1 = d - d_0$ and $d_r = n - d$.

This is exactly the $F$-test from the canonical model: the ratio of scaled signal magnitude to scaled residual magnitude. It measures whether the improvement in fit from adding the extra predictors is large relative to the residual noise.

### 4.5 Individual Coefficient $t$-Test and Confidence Interval

A common special case: test $H_0: \beta_j = 0$ for a single coefficient. Here $d_1 = 1$.

**Geometric interpretation.** The signal direction for testing $\beta_j = 0$ is the part of column $X_j$ that is orthogonal to all the other columns of $X$ — the **unique contribution** of predictor $j$ beyond the other predictors. The $t$-statistic measures how much $Y$ projects onto this direction, relative to the residual noise.

**Test statistic:**
$$T = \frac{\hat\beta_j}{\widehat{\text{SE}}(\hat\beta_j)} \sim t_{n - d} \quad \text{under } H_0,$$
where $\widehat{\text{SE}}(\hat\beta_j) = \hat\sigma \sqrt{[(X^T X)^{-1}]_{jj}}$ is the estimated standard error of $\hat\beta_j$.

This is the $d_1 = 1$ case of the canonical $t$-test — the same construction as the one-sample $t$ from Lecture 23 and the two-sample $t$ from §3, in yet another coordinate system.

**Confidence interval** (by test inversion, as in §1.4):

$$\hat\beta_j \pm t_{n-d}(\alpha/2) \cdot \widehat{\text{SE}}(\hat\beta_j).$$

This is the standard regression confidence interval that appears in every software output. It comes from inverting the canonical $t$-test: the set of values $\beta_j^0$ for which $|\hat\beta_j - \beta_j^0| / \widehat{\text{SE}}(\hat\beta_j) \leq t_{n-d}(\alpha/2)$.

## 5. Summary

| Concept | Key Idea |
|---|---|
| **Canonical LM** | Three blocks $(d_0, d_1, d_r)$; four tests ($z, \chi^2, t, F$); CIs by inversion. |
| **Rotation recipe** | Build $Q = [Q_0\ \|\ Q_1\ \|\ Q_r]$ from orthonormal bases for $\Theta_0$, $\Theta \cap \Theta_0^\perp$, $\Theta^\perp$. Apply $Z = Q^T Y$ to reach canonical form. |
| **Two-sample $t$** | $\Theta = \text{Span}(\mathbf a, \mathbf b)$, $\Theta_0 = \text{Span}(\mathbf 1)$. $(d_0, d_1, d_r) = (1, 1, m+n-2)$. $T = (\bar X - \bar Y)/(S_p \sqrt{1/m + 1/n}) \sim t_{m+n-2}$. |
| **Regression $F$-test** | $F = (\text{RSS}_0 - \text{RSS})/d_1 \div \text{RSS}/d_r \sim F_{d_1, d_r}$. Measures whether tested predictors improve the fit. |
| **Individual $t$-test** | $T = \hat\beta_j / \widehat{\text{SE}}(\hat\beta_j) \sim t_{n-d}$. Signal = unique contribution of predictor $j$. CI: $\hat\beta_j \pm t_{n-d}(\alpha/2) \cdot \widehat{\text{SE}}(\hat\beta_j)$. |

**Key takeaways.**

1. The **canonical linear model** with block dimensions $(d_0, d_1, d_r)$ is a universal template: four test statistics ($z$, $\chi^2$, $t$, $F$), confidence intervals by inversion, all driven by ratios of projection magnitudes.

2. The **rotation recipe** converts any linear-model testing problem to canonical form: build $Q = [Q_0\ |\ Q_1\ |\ Q_r]$ from orthonormal bases for $\Theta_0$, $\Theta \cap \Theta_0^\perp$, and $\Theta^\perp$.

3. **Two-sample $t$**: $(d_0, d_1, d_r) = (1, 1, m+n-2)$. The signal direction is the contrast between group means. The pooled variance estimate uses residuals from both groups.

4. **Regression $F$-test**: compares $(\text{RSS}_0 - \text{RSS}) / d_1$ to $\text{RSS} / d_r$. The individual-coefficient $t$-test and CI ($\hat\beta_j \pm t_{n-d}(\alpha/2) \cdot \widehat{\text{SE}}(\hat\beta_j)$) are the $d_1 = 1$ special case.

5. All of these tests rest on **one fact**: rotational invariance of $N_n(0, \sigma^2 I_n)$, which gives independence of projections onto orthogonal subspaces.
